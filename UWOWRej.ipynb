```python
if {"DATE", "TIME"}.issubset(df.columns):
    df["timestamp"] = pd.to_datetime(df["DATE"] + " " + df["TIME"])
else:
    df["timestamp"] = pd.to_datetime(df["timestamp"]) 
LSL = 580     
USL = 700      
y = "DATA1"    

under = df[df[y] < LSL]
over  = df[df[y] > USL]

plt.figure(figsize=(12, 6))

plt.plot(df["timestamp"], df[y], lw=0.8, label="Weight", color="lightgreen")

plt.axhspan(LSL, USL, color="lightgreen", alpha=0.25,
            label=f"Inspec band ({LSL}{USL}g)")

plt.scatter(under["timestamp"], under[y], color="red",
            s=18, label="Underweight (<580g)")
plt.scatter(over["timestamp"], over[y],  color="orange",
            s=18, label="Overweight (>700g)")

plt.xlabel("Timestamp")
plt.ylabel("Weight (g)")
plt.title("Cake Weight vs. Time with Spec Band")
plt.legend(loc="upper right")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Aastha\Downloads\UW-OW 12th-20th June.csv")

df_clean = df.dropna(subset=['L', 'C', 'R']).copy()

df_clean[['L', 'C', 'R']] = df_clean[['L', 'C', 'R']].astype(float)

df_clean['avg_thickness'] = df_clean[['L', 'C', 'R']].mean(axis=1)

df_clean['thickness_status'] = df_clean['avg_thickness'].apply(
    lambda x: 'Underweight' if x < 1.133 else ('Overweight' if x > 1.163 else 'Ideal')
)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

uw_data = df_clean[df_clean['thickness_status'] == 'Underweight']
sns.regplot(ax=axes[0], x='avg_thickness', y='UW', data=uw_data,
            scatter_kws={"color": "blue"}, line_kws={"color": "red"})
axes[0].set_title('Underweight Rejections vs Average Thickness')
axes[0].set_xlabel('Average Thickness')
axes[0].set_ylabel('Underweight Rejections')

ow_data = df_clean[df_clean['thickness_status'] == 'Overweight']
sns.regplot(ax=axes[1], x='avg_thickness', y='OW', data=ow_data,
            scatter_kws={"color": "green"}, line_kws={"color": "red"})
axes[1].set_title('Overweight Rejections vs Average Thickness')
axes[1].set_xlabel('Average Thickness')
axes[1].set_ylabel('Overweight Rejections')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Aastha\Downloads\UW-OW 12th-20th June.csv")

df_clean = df.dropna(subset=['L', 'C', 'R']).copy()

df_clean[['L', 'C', 'R']] = df_clean[['L', 'C', 'R']].astype(float)

df_clean['avg_thickness'] = df_clean[['L', 'C', 'R']].mean(axis=1)


df_clean.set_index('Timestamp', inplace=True)

df_clean['thickness_deviation'] = df_clean['avg_thickness'] - 1.15

df_clean['UW_due_to_thickness'] = (df_clean['thickness_deviation'] < (-0.015))
df_clean['OW_due_to_thickness'] = (df_clean['thickness_deviation'] > 0.015) 

plt.figure(figsize=(14, 6))
plt.plot(df_clean.index, df_clean['UW'], label='Underweight Rejections', color='blue')
plt.scatter(df_clean.index[df_clean['UW_due_to_thickness']],
            df_clean['UW'][df_clean['UW_due_to_thickness']],
            color='red', label='UW due to Thickness', zorder=5)
plt.title("Underweight Rejections Over Time")
plt.xlabel("Time")
plt.ylabel("UW Rejections")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ---- Overweight Time Series Plot ----
plt.figure(figsize=(14, 6))
plt.plot(df_clean.index, df_clean['OW'], label='Overweight Rejections', color='green')
plt.scatter(df_clean.index[df_clean['OW_due_to_thickness']],
            df_clean['OW'][df_clean['OW_due_to_thickness']],
            color='orange', label='OW due to Thickness', zorder=5)
plt.title("Overweight Rejections Over Time")
plt.xlabel("Time")
plt.ylabel("OW Rejections")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
